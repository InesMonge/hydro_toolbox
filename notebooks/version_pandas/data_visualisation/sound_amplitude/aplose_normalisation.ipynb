```python
# Normalised proportion of labels occurrences (different from annotation count)
%matplotlib qt
# Extraction of spectrograms id in the dataset
all_spectro_ids = df['filename'].str.extract(r'(\d{6})(?=\.wav)')[0].unique()
total_spectro_count = len(all_spectro_ids)

labels = df['label'].unique().tolist()
proportions_global = {}

for label in labels:
    # Filter DataFrame by label
    df_label = df[df['label'] == label].copy()
    label_spectro_ids = df_label['filename'].str.extract(r'(\d{6})(?=\.wav)')[0].unique()
    proportion = len(label_spectro_ids) / total_spectro_count
    proportions_global[label] = proportion

# Sort labels by occurrences
sorted_items = sorted(proportions_global.items(), key=lambda x: x[1])
sorted_labels = [item[0] for item in sorted_items]
sorted_proportions = [item[1] for item in sorted_items]

# # Colormap
# palette = sns.color_palette("tab20", n_colors=20)

# Plot bar chat
plt.style.use('default')
plt.figure(figsize=(10,6))
plt.bar(sorted_labels, sorted_proportions, color='darkcyan')
plt.xticks(rotation=90)
plt.ylabel('Normalised occurrence of labels across spectrograms')
plt.grid(True, alpha=0.5, linestyle='--', color='gray')
plt.tight_layout()
plt.show()

# plt.savefig(os.path.join('/home/imonge/Documents/annotations', f'normalised_proportion_of_label.png'))
```

```python
# WITH confidence indicator - 0/1
df['start_datetime'] = pd.to_datetime(df['start_datetime']).dt.tz_localize(None)

date_start = pd.to_datetime('2021-08-01')
date_end = pd.to_datetime('2024-09-01')
full_index = pd.date_range(start=date_start, end=date_end, freq='MS')

labels = df['label'].unique().tolist()

for label in labels:
    # Filter DataFrame by label
    df_label = df[df['label'] == label].copy()

    # extract month
    df_label['month'] = df_label['start_datetime'].dt.to_period('M').dt.to_timestamp()

    # Extract spectro ID from filename
    df_label['spectro_id'] = df_label['filename'].str.extract(r'(\d{6})(?=\.wav)')

    # Group by month and spectro to classify confidence
    grouped = df_label.groupby(['month', 'spectro_id'])

    def classify_conf(group):
        # if at least one 'Sure' -> spectro 'Sure'
        if (group['confidence'] == 'Sure').any():
            return 'Sure'
        # if all 'Not sure' -> spectro 'Not sure'
        elif (group['confidence'] == 'Not sure').all():
            return 'Not sure'
        else:
            return 'Unknown'

    # Apply classification
    confidence_per_spectro = grouped.apply(classify_conf).reset_index(name='final_confidence')

    #Total number of unique spectros per month (regardless of label)
    df['month'] = df['start_datetime'].dt.to_period('M').dt.to_timestamp()
    df['spectro_id'] = df['filename'].str.extract(r'(\d{6})(?=\.wav)')
    total_spectros = df[['month', 'spectro_id']].drop_duplicates().groupby('month').size()

    # Count Sure / Not sure spectros by month
    presence_conf = confidence_per_spectro.groupby(['month', 'final_confidence']).size().unstack(fill_value=0)

    for col in ['Sure', 'Not sure']:
        if col not in presence_conf.columns:
            presence_conf[col] = 0
    presence_conf = presence_conf[['Not sure', 'Sure']]

    # Normalisation
    presence_conf =  presence_conf.reindex(total_spectros.index)
    norm_presence = (presence_conf.T / total_spectros).T
    norm_presence = norm_presence.reindex(full_index, fill_value=0) # To have all month (with or without label annotated for the month)


    # Plot
    plt.style.use('default')
    fig, ax = plt.subplots(figsize=(12,6))

    ax.bar(norm_presence.index, norm_presence['Sure'].values, width=20, align='center', label='Sure', color='darkcyan')
    ax.bar(norm_presence.index, norm_presence['Not sure'].values, width=20, align='center', label = 'Not sure', color='mediumaquamarine', bottom=norm_presence['Sure'].values)
    # norm_presence.plot(kind='bar', ax=ax, stacked=True, color=['mediumaquamarine', 'darkcyan'], width=20)

    # Titles
    ax.set_title(f'Presence of {labels_info[label]} by month')
    ax.set_xlabel('Date')
    ax.set_ylabel('Proportion of files with detection (0-1)')

    # Plot margins
    start_margin = full_index.min() - pd.DateOffset(day=15)
    end_margin = full_index.max() + pd.DateOffset(day=15)
    ax.set_xlim(start_margin, end_margin)
    ax.set_ylim(0, 1.05)

    # Ticks parameters
    # Major
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
    ax.tick_params(axis='x', which='major', bottom=True, top=False, length=12, width=1.5, color='black')
    # Minor
    ax.xaxis.set_minor_locator(mdates.MonthLocator())
    ax.tick_params(axis='x', which='minor', bottom=True, top=False, length=5, color='black')

    ax.legend(title='Confidence', loc='upper right')
    ax.grid(True, alpha=0.4, linestyle='--', color='gray')
    plt.tight_layout()
    plt.savefig(os.path.join(output_path, 'normalised_0-1_seasonality_confidence_new', f'normalised_0-1_seasonality_confidence_{label}_new.png'))
    # plt.show()
    plt.close()
```
