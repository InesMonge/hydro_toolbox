```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
df_info = pd.read_pickle('../../../../../data/SPL/df_segments_updated.pkl')
# df_psd = pd.read_pickle('../../../../../data/SPL/psd_aplose_fq_filtered.pkl')
df_psd = pd.read_pickle('../../../../../data/SPL/psd_aplose_fq_filtered_no_background.pkl') # without background noise
output_path = "/home/imonge/Documents/annotations"
# Delete columns 'signal' and 'sampling_rate'
df_info = df_info.drop(columns=['signal', 'sampling_rate'], errors='ignore')
```

```python
# Reset indexes for both datasets before merging
df_info = df_info.reset_index(drop=True)
df_psd = df_psd.reset_index(drop=True)

# Merge both dataframes line by line
df_merged = pd.concat([df_info, df_psd], axis=1)

# Post-merge checking
assert df_merged.shape[0] == df_info.shape[0] == df_psd.shape[0], "Allignment error after merging"

labels = df_merged['label'].unique()
```

```python
# Plot labels individually with mean PSD on the 4-year data
for label in labels:
    df_label = df_merged[df_merged['label'] == label]

    # Only take species (no geophony or anthropophony)
    if df_label['species'].replace('', np.nan).notna().sum() == 0:
        continue

    # Extract only frequency columns
    df_label_psd = df_label[df_psd.columns]

    # Mean PSD by label ###
    mean_psd = df_label_psd.mean(axis=0)

    # Convert to dB re 1 µPa²/Hz
    mean_psd[mean_psd < 1e-12] = np.nan # avoid log(0), put Nan instead
    mean_psd_db = 10 * np.log10(mean_psd)

    plt.figure(figsize=(12,6))
    plt.plot(mean_psd_db.index, mean_psd_db.values, label=label, color='darkcyan')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Mean amplitude (dB re 1 µPa²/Hz)') ## Chechk if it is in fact 1µPa²
    plt.title(f'Peak frequencies in the annotation box - {label}')
    plt.style.use('default')
    plt.grid(True, alpha=0.4, linestyle='--')
    plt.tight_layout()
    plt.show()

    # plt.savefig(os.path.join(output_path, 'peak_fq_no_background', f'peak_frequency_no_background_{label}.png'))
    # plt.close()
```

```python
# Plot all labels
cmap = plt.get_cmap('tab20')
colors = {label: cmap(i % cmap.N) for i, label in enumerate(labels)}


plt.style.use('default')
fig, ax = plt.subplots(figsize=(14,8))

for label in labels:
    df_label = df_merged[df_merged['label'] == label]

    # Only take species (no geophony or anthropophony)
    if df_label['species'].replace('', np.nan).notna().sum() == 0:
        continue

    # Extract only frequency columns
    df_label_psd = df_label[df_psd.columns]

    # Mean PSD by label
    mean_psd = df_label_psd.mean(axis=0)

    # Convert to dB re 1 µPa²/Hz
    mean_psd[mean_psd < 1e-12] = np.nan # avoid log(0), put Nan instead
    mean_psd_db = 10 * np.log10(mean_psd)

    ax.plot(mean_psd_db.index, mean_psd_db.values, label=label, color=colors[label])

ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Mean amplitude (dB re 1 µPa²/Hz)')
ax.set_xlim(13, 120)
ax.set_ylim(55, 102)
ax.set_title('Mean spectra')
ax.legend()
plt.grid(True, alpha=0.4, linestyle='--')
plt.tight_layout()
# plt.show()

plt.savefig(os.path.join(output_path, f'peak_frequency_all_no_background.png'))
plt.close()
```
