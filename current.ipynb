# **Current**
Correlations between **current speed** *(√(uo^2+vo^2)* and the mean amplitude for different frequencies

Initialisation

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import sys
import xarray as xr

from scipy.signal import medfilt
from utils.data_reading.sound_data.station import StationsCatalog
```

```python
sys.path.append("/home/imonge/PycharmProjects/PythonProject/toolbox/src") # Path of the root directory
catalog_path = "/media/imonge/CORSAIR" # Path of the hard disk containing the MAHY data

stations = StationsCatalog(catalog_path)
# %matplotlib qt #-> Use this in case you want to have the plot in a new window
```

```python
# Take the time dates from the MAHY csv
cat_info = pd.read_csv(catalog_path + '/MAHY/MAHY.csv', parse_dates=['date_start', 'date_end'])
cat_info.set_index('station_name', inplace=True)
cat_info["date_start"] = pd.to_datetime(cat_info["date_start"])
cat_info["date_end"] = pd.to_datetime(cat_info["date_end"])

# Determine the first date and last date of all MAHY campaigns to match with the current speed dates
date_start_total = cat_info["date_start"].min()
date_end_total = cat_info["date_end"].max()
```

Save current data in a numpy file

```python
output_path_current = "/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_soundscape/current" # Path of the output directory for the current data

for station in stations:

    hydro_number = station.name[-1]
    station_code = f"h{hydro_number}"

    if "depth" in station.other_kwargs:
            depth_choice = station.other_kwargs["depth"]  # récupère la colonne depth des csv
    else:
        depth_choice = 1300

    date_start = date_start_total
    date_end = date_end_total

    current_path = f"/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_copernicus/COPER_MAHY*{hydro_number}.nc" # Path of the directory containing the current data for one hydrophone
    current_data = xr.open_dataset(current_path)

    t = pd.to_datetime(current_data["time"].values)

    mask = (t>date_start) & (t<date_end)

    depth_index = np.argmin(abs(current_data["depth"].data-depth_choice))

    # Calculate the current speed
    uo = np.array(current_data["uo"][:,depth_index]).flatten()
    vo = np.array(current_data["vo"][:,depth_index]).flatten()
    current_speed = np.sqrt(uo**2 + vo**2)[mask]

    # Save the current speed data as a numpy file
    np.save(os.path.join(output_path_current, f"current_speed_" + station_code + ".npy"), current_speed)
```

Choose the stations

```python
# Select the stations you want to plot. /!\ You can only choose stations for one hydrophone
selected_stations = ["MAHY02",
                     "MAHY12",
                     "MAHY22",
                     "MAHY32",
                     "MAHY42"
                     ]

hydrophone_number = [station[-1] for station in selected_stations]
# Check that only one hydrophone is chosen
if len(set(hydrophone_number))>1:
    raise ValueError("You have selected more than one hydrophone")
else:
    print(f"Hydrophone n°{hydrophone_number[0]} selected")
```

Spectrogram with current speed overlaid for the stations selected

```python
# Path of the directories containing psds and current speed data
mahy_path = "/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_mahy"
current_path = f"/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_soundscape/current/current_speed_h{hydrophone_number[0]}.npy"

# Take the dates from de MAHY csv
date_start = cat_info["date_start"].loc[selected_stations[0]]
date_end = cat_info["date_end"].loc[selected_stations[-1]]

# Spectrogram
psds_all = []

for station in selected_stations:
    psd_path = os.path.join(mahy_path, f"PSDS_{station}.npy")
    psds = np.load(psd_path)
    psds_all.append(psds)

# Concatenate psds
final_psd = np.vstack(psds_all)

# Current speed
current_speed = np.load(current_path)

dates_all = pd.date_range(start=date_start, end=date_end, periods=len(current_speed))
assert len(current_speed) == len(dates_all); "Mismatch in current_speed and dates_all length"

# Interpolate current speed in order to correspond to the spectrogram dates
if len(dates_all) != len(current_speed):
    current_speed_interp = np.interp(pd.to_numeric(dates_all), pd.to_numeric(dates_all), current_speed)

else:
    current_speed_interp = current_speed

# Create plot and axis
fig, ax = plt.subplots(figsize = (16,10))
ax1 = ax.twinx() # Second axis for current speed

sp = ax.imshow(10*np.log10(final_psd.T + 1e-12), aspect="auto", origin="lower", cmap="viridis", extent=(date_start, date_end, 0, 120), vmin=10, vmax=95)

# Trace current speed on the second axis
ax1.plot(dates_all, current_speed_interp,label="Current speed (m/s)", color="blue")
ax1.set_ylim(0,0.4) #limites ax1
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.grid(False, axis='y') # No grid for the frequency

# Add labels
ax.set_xlabel("Time")
ax.set_ylabel("Frequency (Hz)")
ax1.set_ylabel("Current speed (m/s)")
ax1.legend(loc="upper left")
ax.set_title("Current speed (m/s) overlaid on the spectrogram")
```

Correlations between the median amplitude and the current speed as a function of frequency

```python
# Define the amplitude correlations
amplitude_corr = []

# Calculate the correlation between the filtered amplitude and the interpolated current
for i in range(final_psd.shape[1]):
    amplitude = final_psd[:, i]
    amplitude_filt = medfilt(amplitude, kernel_size=51)

    amplitude_steps = amplitude_filt.shape[0]
    t_amplitude = pd.date_range(
        start=date_start,
        end=date_end,
        periods=amplitude_steps)

    #interpolated_current = interp1d(pd.to_numeric(dates_all), current_speed, bounds_error=False)(pd.to_numeric(t_amplitude))
    interpolated_current = np.interp(pd.to_numeric(t_amplitude), pd.to_numeric(dates_all), current_speed)
    validmask = (~np.isnan(interpolated_current)) & (~np.isnan(amplitude_filt) & (amplitude_filt != 0)) # remove nan values and null values

    if np.sum(validmask) > 0:
        corr_matrix = np.corrcoef(interpolated_current[validmask], amplitude_filt[validmask])
        amplitude_corr.append(corr_matrix[0, 1])
    else:
        amplitude_corr.append(np.nan)

# Calculate the associated frequencies
fs = 240
nperseg = 1024
frequencies = np.fft.rfftfreq(nperseg, d=1 / fs)

# Create plot
plt.figure(figsize=(16,10))
plt.plot(frequencies, amplitude_corr, color="steelblue", label="Correlation median sound amplitude / current speed")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Pearson correlation coefficient")
plt.title("Correlations between the median amplitude and the current speed as a function of frequency (Hz)")
plt.legend()
plt.tight_layout()
plt.show()
```
