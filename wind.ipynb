# **Wind**
Correlations between **wind speed** (√(u10^2+v10^2) and the mean amplitude for different frequencies

Initialisation

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import sys
import xarray as xr

from scipy.signal import medfilt
from utils.data_reading.sound_data.station import StationsCatalog
```

```python
sys.path.append("/home/imonge/PycharmProjects/PythonProject/toolbox/src") # Path of the root directory
catalog_path = "/media/imonge/CORSAIR" # Path of the hard disk containing the MAHY data

stations = StationsCatalog(catalog_path)
# %matplotlib qt #-> Use this in case you want to have the plot in a new window
```

```python
# Take the time dates from the MAHY csv
cat_info = pd.read_csv(catalog_path + '/MAHY/MAHY.csv', parse_dates=['date_start', 'date_end'])
cat_info.set_index('station_name', inplace=True)
cat_info["date_start"] = pd.to_datetime(cat_info["date_start"])
cat_info["date_end"] = pd.to_datetime(cat_info["date_end"])

# Determine the first date and last date of all MAHY campaigns to match with the current speed dates
date_start_total = cat_info["date_start"].min()
date_end_total = cat_info["date_end"].max()
```

Save wind data in a numpy file

```python
output_path_wind = "/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_soundscape/wind"

for station in stations:

    hydro_number = station.name[-1]
    station_code = f"h{hydro_number}"

    date_start = date_start_total
    date_end = date_end_total

    wind_path = f"/home/imonge/PycharmProjects/PythonProject/toolbox/src/ERA5_concat/ERA5_MAHY*{hydro_number}_wind_merged.nc"
    wind_data = xr.open_dataset(wind_path)

    t = pd.to_datetime(wind_data["valid_time"].values)

    mask = (t>date_start) & (t<date_end)

    u10 = np.array(wind_data["u10"]).flatten()
    v10 = np.array(wind_data["v10"]).flatten()
    wind_speed = np.sqrt(u10**2 + v10**2)[mask]

    np.save(os.path.join(output_path_wind, f"wind_speed_" + station_code + ".npy"), wind_speed)
```

Spectrogram with wind speed overlaid for the stations selected

```python
# Path of the directories containing psds and wind speed data
mahy_path = "/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_mahy"
wind_path = f"/home/imonge/PycharmProjects/PythonProject/toolbox/src/data_soundscape/wind/wind_speed_h{hydrophone_number[0]}.npy"

# Take the dates from de MAHY csv
date_start = cat_info["date_start"].loc[selected_stations[0]]
date_end = cat_info["date_end"].loc[selected_stations[-1]]

# Spectrogram
psds_all = []

for station in selected_stations:
    psd_path = os.path.join(mahy_path, f"PSDS_{station}.npy")
    psds = np.load(psd_path)
    psds_all.append(psds)

# Concatenate psds
final_psd = np.vstack(psds_all)

# Wind speed
wind_speed = np.load(wind_path)

dates_all = pd.date_range(start=date_start, end=date_end, periods=len(wind_speed))
assert len(wind_speed) == len(dates_all); "Mismatch in wind_speed and dates_all length"

# Interpolate wind speed in order to correspond to the spectrogram dates
if len(dates_all) != len(wind_speed):
    wind_speed_interp = np.interp(pd.to_numeric(dates_all), pd.to_numeric(dates_all), wind_speed)

else:
    wind_speed_interp = wind_speed

# Gather wind speed values daily
df_wind = pd.DataFrame({"wind_speed":wind_speed_interp}, index=dates_all)
daily_wind = df_wind.resample('D').median()

# Create plot and axis
fig, ax = plt.subplots(figsize = (16,10))
ax1 = ax.twinx() # Second axis for wind speed

sp = ax.imshow(10*np.log10(final_psd.T + 1e-12), aspect="auto", origin="lower", cmap="viridis", extent=(date_start, date_end, 0, 120), vmin=10, vmax=95)

# Trace wind speed on the second axis
ax1.plot(daily_wind.index, daily_wind['wind_speed'],label="Wind speed (m/s) - Daily Median", color="blue")
ax1.set_ylim(0,20) #limites ax1
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.grid(False, axis='y') # No grid for the frequency

# Add labels
ax.set_xlabel("Time")
ax.set_ylabel("Frequency (Hz)")
ax1.set_ylabel("Wind speed (m/s) - Daily Median")
ax1.legend(loc="upper left")
ax.set_title("Daily median wind speed (m/s) overlaid on the spectrogram")
```

Correlations between the median amplitude and the wind speed as a function of frequency

```python
# Define the amplitude correlations
amplitude_corr = []

# Calculate the correlation between the filtered amplitude and the interpolated wind
for i in range(final_psd.shape[1]):
    amplitude = final_psd[:, i]
    amplitude_filt = medfilt(amplitude, kernel_size=51)

    amplitude_steps = amplitude_filt.shape[0]
    t_amplitude = pd.date_range(
        start=date_start,
        end=date_end,
        periods=amplitude_steps)

    #interpolated_wind= interp1d(pd.to_numeric(dates_all), wind_speed, bounds_error=False)(pd.to_numeric(t_amplitude))
    interpolated_wind = np.interp(pd.to_numeric(t_amplitude), pd.to_numeric(dates_all), wind_speed)
    validmask = (~np.isnan(interpolated_wind)) & (~np.isnan(amplitude_filt) & (amplitude_filt != 0)) # remove nan values and null values

    if np.sum(validmask) > 0:
        corr_matrix = np.corrcoef(interpolated_wind[validmask], amplitude_filt[validmask])
        amplitude_corr.append(corr_matrix[0, 1])
    else:
        amplitude_corr.append(np.nan)

# Calculate the associated frequencies
fs = 240
nperseg = 1024
frequencies = np.fft.rfftfreq(nperseg, d=1 / fs)

# Create plot
plt.figure(figsize=(16,10))
plt.plot(frequencies, amplitude_corr, color="steelblue", label="Correlation median sound amplitude / wind speed")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Pearson correlation coefficient")
plt.title("Correlations between the median amplitude and the wind speed as a function of frequency (Hz)")
plt.legend()
plt.tight_layout()
plt.show()
```
